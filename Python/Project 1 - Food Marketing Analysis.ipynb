```python
age_groups = [(23,30),(31,40),(41,50),(51,60),(61,70),(71,85)]

def assign_age_group(Age):
    for age_range in age_groups:
        if age_range[0] <= Age <= age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return("Unknown")

food['Age_Group'] = food['Age'].apply(assign_age_group)
```

# OVERALL FINDINGS

### 1. Age Insight:  Customers aged 3070 contribute higher overall spending, but are less likely to accept campaigns, although they represent a high-volume customer segment.

### 2. Channel Insight: Catalog campaigns show higher acceptance rates, while in-store purchases generate higher spending. A balanced channel strategy is recommended: 40% catalog, 30% store, 30% web.

### 3. Family Insight: Customers with no children or fewer children are more likely to accept campaigns and should be prioritized in targeting strategies.

### 4. Education Insight: Education level shows minimal impact on spending or campaign acceptance and should not be a primary targeting factor

### 5. Marital Status Insight: Marital status has limited influence on campaign acceptance; however, Married, Single, and Together customers contribute higher overall spending.

### Revenue Strategy:

### 1. Focus on middle-aged, high-income customers with no children.
### 2. Target this segment across multiple platforms using the recommended channel split.


### Customer Acquisition:
### 1. Focus on younger (2130) and older (70+) customers, as these age groups show higher campaign acceptance rates and potential for future growth.
